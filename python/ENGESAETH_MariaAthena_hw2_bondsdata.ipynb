```python
# Compute the K-Means cluster for the Corporate Bonds Data Set

# K-Means for k=2
n_clusters2=2
k_means2 = cluster.KMeans(n_clusters=n_clusters2, init='k-means++')
k_means2.fit(c_data) # Computes k-means

k_means_labels2 = k_means2.labels_
centroids2 = k_means2.cluster_centers_
k_means_labels_unique2 = np.unique(k_means_labels2)


# K-Means for k=3
n_clusters3=3
k_means3 = cluster.KMeans(n_clusters=n_clusters3, init='k-means++')
k_means3.fit(c_data) # Computes k-means

k_means_labels3 = k_means3.labels_
centroids3 = k_means3.cluster_centers_
k_means_labels_unique3 = np.unique(k_means_labels3)
```

```python
# Plot result
fig = plt.figure(figsize=(18,6), dpi=1600)
#fig.subplots_adjust(left=0.02, right=0.98, bottom=0.05, top=0.9)

colors2 = ['#4EACC5', '#FF9C34']
colors3 = ['#4EACC5', '#FF9C34', '#4E9A06']


# KMeans, k=2
ax2 = fig.add_subplot(1, 2, 1) # rows, columns, plot nb.
for k, col in zip(range(n_clusters2), colors2):
    my_members2 = k_means_labels2 == k
    cluster_center2 = centroids2[k]
    ax2.plot(c_data[my_members2, 0], c_data[my_members2, 1], 'w',
            markerfacecolor=col, marker='.')
    ax2.plot(cluster_center2[0], cluster_center2[1], 'o', markerfacecolor='grey',
            markeredgecolor='w', markersize=20)
ax2.set_title('K-Means Clustering for Corporate Bonds dataset \n'
             'k=2, centroids are marked with grey circles')
plt.xlabel("Coupon Rate")
plt.ylabel("YTM")


# KMeans, k=3
ax3 = fig.add_subplot(1, 2, 2)
for k, col in zip(range(n_clusters3), colors3):
    my_members3 = k_means_labels3 == k
    cluster_center3 = centroids3[k]
    ax3.plot(c_data[my_members3, 0], c_data[my_members3, 1], 'w',
            markerfacecolor=col, marker='.')
    ax3.plot(cluster_center3[0], cluster_center3[1], 'o', markerfacecolor='grey',
            markeredgecolor='w', markersize=20)
ax3.set_title('K-Means Clustering for Corporate Bonds dataset \n'
             'k=3, centroids are marked with grey circles')
plt.xlabel("Coupon Rate")
plt.ylabel("YTM")

#plt.show()
```

### Q3: Visualise Mean Silhouette Score

#### Corporate Bonds Dataset

```python
# Convert the two features Coupon and YTM to numpy array
c_data = np.array(corp_df[['Coupon', 'YTM']])
data = c_data

# Loop through values for k ranging from 1 to 10 to calcualte the silhouette score
range_n_clusters = [2, 3, 4, 5, 6, 7, 8, 9, 10]

for n_clusters in range_n_clusters:
    # Create a subplot with 1 row and 2 columns
    fig, (ax1, ax2) = plt.subplots(1, 2)
    fig.set_size_inches(18, 7)

    # The 1st subplot is the silhouette plot
    # The silhouette coefficient can range from -1, 1 but in this example all
    # lie within [-0.1, 1]
    ax1.set_xlim([-0.1, 1])
    # The (n_clusters+1)*10 is for inserting blank space between silhouette
    # plots of individual clusters, to demarcate them clearly.
    ax1.set_ylim([0, len(data) + (n_clusters + 1) * 10])
    
    # Initialize the clusterer with n_clusters value and a random generator
    # seed of 10 for reproducibility.
    clusterer = sklearn.cluster.KMeans(n_clusters=n_clusters, random_state=10)
    cluster_labels = clusterer.fit_predict(data)

    # The silhouette_score gives the average value for all the samples.
    # This gives a perspective into the density and separation of the formed
    # clusters
    silhouette_avg = sklearn.metrics.silhouette_score(data, cluster_labels)

    # Compute the silhouette scores for each sample
    sample_silhouette_values = silhouette_samples(data, cluster_labels)
    
    y_lower = 10

    for i in range(n_clusters):
        # Aggregate the silhouette scores for samples belonging to
        # cluster i, and sort them
        ith_cluster_silhouette_values = \
            sample_silhouette_values[cluster_labels == i]

        ith_cluster_silhouette_values.sort()

        size_cluster_i = ith_cluster_silhouette_values.shape[0]
        y_upper = y_lower + size_cluster_i

        color = cm.spectral(float(i) / n_clusters)
        ax1.fill_betweenx(np.arange(y_lower, y_upper),
                          0, ith_cluster_silhouette_values,
                          facecolor=color, edgecolor=color, alpha=0.7)

        # Label the silhouette plots with their cluster numbers at the middle
        ax1.text(-0.05, y_lower + 0.5 * size_cluster_i, str(i))

        # Compute the new y_lower for next plot
        y_lower = y_upper + 10  # 10 for the 0 samples

    ax1.set_title("The silhouette plot for the various clusters.")
    ax1.set_xlabel("The silhouette coefficient values")
    ax1.set_ylabel("Cluster label")

    # The vertical line for average silhoutte score of all the values
    ax1.axvline(x=silhouette_avg, color="red", linestyle="--")

    ax1.set_yticks([])  # Clear the yaxis labels / ticks
    ax1.set_xticks([-0.1, 0, 0.2, 0.4, 0.6, 0.8, 1])

    # 2nd Plot showing the actual clusters formed
    colors = cm.spectral(cluster_labels.astype(float) / n_clusters)
    ax2.scatter(data[:, 0], data[:, 1], marker='.', s=30, lw=0, alpha=0.7,
                c=colors)

    # Labeling the clusters
    centers = clusterer.cluster_centers_
    # Draw white circles at cluster centers
    ax2.scatter(centers[:, 0], centers[:, 1],
                marker='o', c="white", alpha=1, s=200)

    for i, c in enumerate(centers):
        ax2.scatter(c[0], c[1], marker='$%d$' % i, alpha=1, s=50)

    ax2.set_title("The visualization of the clustered data.")
    ax2.set_xlabel("Feature space for the 1st feature")
    ax2.set_ylabel("Feature space for the 2nd feature")

    plt.suptitle(("Silhouette analysis for KMeans clustering on sample data "
                  "with n_clusters = %d" % n_clusters),
                 fontsize=14, fontweight='bold')
    
    plt.xlabel("Silhouette Coefficient Values")
    plt.ylabel("Cluster Label")
    plt.show()
```

### Q5: Interpreatation and Analysis of Clustering
##### incl. Bonus Questions

#### Interpreting The Silhouette Score

In Q2 is the numerical output of the computed Silhouette Coefficient of all samples on both datasets.

The Silhouette Coefficient is calculated as follows: [(2)-(1)]/max{(1); (2)}
Where (1) and (2) are given by:
    1. mean intra-cluster distance
    2. mean nearest-cluster distance

The score range is [-1 ; 1] and can be interpreted as follows:
    * ~ -1 : negative values indicate that a sample has been assigned to the wrong cluster
    * ~  0 : samples are close to the decision boundary of which cluster they should be assigned to
    * ~ +1 : samples are far away from the neighboring clusters


#### Our Results

Given that a higher Silhouette Score is more desireable. The results for our datasets, given we have used the euclidean distance metric, indicate that 2 is the optimal value of clusters for both Corporate and Municipal Bonds.

The thickness of the silhouette plot visualise the relative size of the clusters. A similar conclusion can be reached from looking at the silhouette plots that show:
    * n_clusters > 3 should be avoided for the Corporate Bonds dataset
    * n_clusters > 4 should be avoided for the Municipal Bonds dataset
due to the presence of clusters with below average and sometimes negative silhouette scores and also due to wide fluctuations in the size of the silhouette plots.


#### Optimal Solution

How many clusters do you have in your data is an ill-posed question on its own. It depends on what you want to do with the data. Given the nature of our data, and the features used to cluster the it in the above analysis, it makes practical sense to give investors options in terms of risk and return. This combined with the the above analysis I would assume the best answer to number of of clusters to be 3 for the Corporate and Municipal Bonds. Allowing investors to easily self select the level of risk they are willing to take on.

K-means is a TOP-DOWN strategy that allows us to recursively partitions the dataset into a tree of clusters. The algorithm is fast, but does not guarantee that similar instances end up in the same cluster. We continue with a BOTTOM-UP approach in Part II below.
